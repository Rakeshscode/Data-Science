## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

###Playing with Discrete Distributions

**Task9**:

Plot the pdfs (techincally the pmfs) of the following distributions: $Bernoulli(p=0.7)$, $Binomial(n=10,p=0.7)$, $Poisson(λ=7)$, $Geometric(p=0.7)$, $HyperGeometric(N=50,k=35,n=15)$, $Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4])$

Additionally, compute the first four moments for all those distributions--except the Categorical distribution.

````python
pz.distributions.bernoulli.Bernoulli(p=0.7).plot_pdf(support=(0,5))
pz.distributions.binomial.Binomial(n=10, p=0.7).plot_pdf(support=(0,5))
pz.distributions.poisson.Poisson(mu=7).plot_pdf(support=(0,5))
pz.distributions.geometric.Geometric(p=0.7).plot_pdf(support=(0,5))
pz.distributions.hypergeometric.HyperGeometric(N=50,k=35,n=15).plot_pdf(support=(0,5))
pz.distributions.categorical.Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4]).plot_pdf()
````

````python
A = pz.distributions.bernoulli.Bernoulli(p=0.7)
B = pz.distributions.binomial.Binomial(n=10, p=0.7)
C = pz.distributions.poisson.Poisson(mu=7)
D = pz.distributions.geometric.Geometric(p=0.7)
E = pz.distributions.hypergeometric.HyperGeometric(N=50,k=35,n=15)
F = pz.distributions.categorical.Categorical(p=[0.05,0.3,0.05,0.1,0.2,0.4])
````

````python
A.mean(), B.mean(), C.mean(), D.mean(), E.mean(), F.mean()
````

````python
A.var(), B.var(), C.var(), D.var(), E.var()
````

````python
A.skewness(), B.skewness(), C.skewness(), D.skewness(), E.skewness()
````

````python
A.kurtosis(), B.kurtosis(), C.kurtosis(), D.kurtosis(), E.kurtosis()
````

**Task10**:

Describe what process each of the previous discrete distributions models, and their support.

Hint: Make sure to read any of the preliZ documentation, the wikipedia articles, or wolfram alpha.


Bernoulli: Models a single trial with success or failure. [0,1]. Max entropy occurs at p =0.5.

Binomial: Models the # of success in n  independent trials

Poisson: mOdels the # of events occuring in a fixed interval

Geometric: Models the # of trials needed to get the first success in independent trials

Hypergeometric: Models the # of successes in a sample without replacement

Categorial: Models a single trial with k outcomes
